```python
def theorem4_rag_demo_a(docs=SAMPLE_DOCS):
    """
    Merge docs about 'ML' or 'networks' then feed to Groq
    """
    ml_docs = [d for d in docs if "ML" in d or "networks" in d or "Generative" in d]
    if ml_docs:
        merged = graph_flow_merger.merge_documents(ml_docs)
        prompt = f"ML-Related Merged:\n{merged}\nSummarize."
        resp = call_groq_llm(prompt)
        print("[Theorem4 RAG Demo A] LLM:\n", resp, "\n")

def theorem4_rag_demo_b(docs=SAMPLE_DOCS):
    """
    Merge docs about 'blockchain' or 'quantum', feed to Groq
    """
    relevant = [d for d in docs if "blockchain" in d.lower() or "quantum" in d.lower()]
    if relevant:
        merged = graph_flow_merger.merge_documents(relevant)
        prompt = f"Blockchain/Quantum Merged:\n{merged}\nExplain synergy/differences."
        resp = call_groq_llm(prompt)
        print("[Theorem4 RAG Demo B] LLM:\n", resp, "\n")
```

```python
theorem4_rag_demo_a(SAMPLE_DOCS)
```

```python
theorem4_rag_demo_b(SAMPLE_DOCS)
```
